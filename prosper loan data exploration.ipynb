```python
#4. What is the distribution of Employement status


plt.figure(figsize=(10, 6))
employment_status_order = df['EmploymentStatus'].value_counts(ascending=True).index
sns.countplot(y='EmploymentStatus', data=df, order=employment_status_order)
plt.title('Distribution of Employment Status')
plt.xlabel('Count')
plt.ylabel('Employment Status')
plt.show()

# Observation:
# 'Employed' is the most common employment status among borrowers, followed by 'Full-time' and 'Self-employed'.
# The frequency decreases for other categories like 'Not available', 'Other', 'Not employed', and 'Retired'.
```

```python
#5. What is the distribution of income range

plt.figure(figsize=(10, 6))
income_range_order = df['IncomeRange'].value_counts(ascending=True).index
sns.countplot(y='IncomeRange', data=df, order=income_range_order)
plt.title('Distribution of Income Range')
plt.xlabel('Count')
plt.ylabel('Income Range')
plt.show()

# Observation:
# The highest number of borrowers fall within the income range of $25,000 to $49,999.
# There are significant numbers of borrowers in the $50,000 to $74,999 and $75,000 to $99,999 income ranges as well.
# This distribution highlights a middle-income demographic as the primary borrower base.
```

```python
# 1. What is the variations loan amount across different employment statuses?

plt.figure(figsize=(10, 6))
sns.boxplot(x='EmploymentStatus', y='LoanOriginalAmount', data=df)
plt.title('Loan Original Amount vs. Employment Status')
plt.xlabel('Employment Status')
plt.ylabel('Loan Original Amount')
plt.xticks(rotation=45)
plt.show()

# Observation:
# The loan amounts vary significantly across different employment statuses. 
# Employed borrowers tend to have higher loan amounts compared to those who are not employed or retired. 
# The self-employed category shows a wider range of loan amounts.

```

```python
#2 How does the borrower's APR vary across different income ranges?

plt.figure(figsize=(10, 6))
sns.boxplot(x='IncomeRange', y='BorrowerAPR', data=df)
plt.title('Borrower APR vs. Income Range')
plt.xlabel('Income Range')
plt.ylabel('Borrower APR')
plt.xticks(rotation=45)
plt.show()

# Observation:
# Borrower APR generally decreases as the income range increases. 
# Borrowers with lower income ranges tend to have higher APRs, suggesting higher risk associated with lower income borrowers.
# The highest income range shows a narrower spread of APRs.
```

```python
#3 How does the loan amount vary across different income ranges?

plt.figure(figsize=(10, 6))
sns.boxplot(x='IncomeRange', y='LoanOriginalAmount', data=df)
plt.title('Loan Original Amount vs. Income Range')
plt.xlabel('Income Range')
plt.ylabel('Loan Original Amount')
plt.xticks(rotation=45)
plt.show()

# Observation:
# Higher income ranges are associated with higher loan amounts. 
# Borrowers in the lowest income range tend to have smaller loans, while those in the highest income range have significantly larger loans.
```

```python
#4.  How does the borrower's APR vary across different employment statuses?

plt.figure(figsize=(10, 6))
sns.boxplot(x='EmploymentStatus', y='BorrowerAPR', data=df)
plt.title('Borrower APR vs. Employment Status')
plt.xlabel('Employment Status')
plt.ylabel('Borrower APR')
plt.xticks(rotation=45)
plt.show()

# Observation:
# Borrower APR varies by employment status, with not employed and self-employed borrowers generally having higher APRs. 
# Employed borrowers have lower APRs, indicating they are considered lower risk by lenders.
```

```python
#5. How does the distribution of loan statuses vary across different employment statuses?

plt.figure(figsize=(10, 6))
sns.countplot(y='LoanStatus', hue='EmploymentStatus', data=df)
plt.title('Loan Status vs. Employment Status')
plt.xlabel('Count')
plt.ylabel('Loan Status')
plt.legend(title='Employment Status', bbox_to_anchor=(1, 1))
plt.show()

# Observation:
# Employed borrowers have a higher number of completed and current loans, indicating better loan performance. 
# Retired and not employed categories show a higher proportion of defaulted and charged-off loans, suggesting higher risk associated with these employment statuses.

```
